```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

90CCDF3FD9A9
```

```python
img = cv2.imread('../data/images/felfel-dolme.jpg')

hue_spectrum = np.linspace(0,179,dtype='uint8',num=180)
img_spec = np.array([hue_spectrum] * 50)
saturation_channel = np.ones_like(img_spec , dtype='uint8')*255
value_channel = np.ones_like(img_spec , dtype='uint8')*255

hsv_spec = cv2.merge([img_spec, saturation_channel, value_channel])
final_spec = cv2.cvtColor(hsv_spec, cv2.COLOR_HSV2BGR)

hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
hue_hist= cv2.calcHist([hsv], [0], None, [180], [0, 180])

plt.figure(figsize=(10, 3))
plt.subplot(1, 2, 1) ; plt.imshow(img[... , ::-1]) ; plt.xticks([]) ; plt.yticks([])
plt.subplot(2, 2, 2) ; plt.plot(hue_hist , color='green') ; plt.xticks([]) ; plt.yticks([])
plt.subplot(2, 2, 4) ; plt.imshow(final_spec[...,::-1]) ; plt.xticks([]) ; plt.yticks([])
```
