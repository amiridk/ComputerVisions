### Binary image processing is a fundamental technique in digital image analysis, where images are represented using only two pixel intensity values, typically 0 (black) and 1 (white). This simplification allows for efficient processing and analysis, particularly in tasks such as object detection, segmentation, and morphological operations.
> By converting grayscale or color images into binary form through thresholding, complex structures can be reduced to their essential shapes, making it easier to identify edges, contours, and regions of interest. Binary image processing is widely used in applications like document scanning, medical imaging, and computer vision, where the focus is on extracting meaningful information from the shape and structure of objects rather than their color or texture. Its computational efficiency and simplicity make it a powerful tool in real-time systems and resource-constrained environments.

```python
# Thresholding with numpy 

import cv2 
import numpy as np 
import matplotlib.pyplot as plt

img = cv2.imread('../data/images/gradient.jpg')
new_img = np.copy(img)
new_img[new_img < 127] = 0
new_img[new_img > 127] = 255

plt.figure(figsize=(10 , 5))
plt.subplot(121);plt.imshow(img , cmap='gray');plt.title('Original')
plt.subplot(122);plt.imshow(new_img , cmap='gray');plt.title('Output')
```

```python
#THRESHOLDING USING OPEN_CV

floor = 127
max = 255
ret , new_img = cv2.threshold(img , floor , max  , cv2.THRESH_BINARY)
print(ret)

plt.figure(figsize=(10 , 5))
plt.subplot(121);plt.imshow(img , cmap='gray');plt.title('Original')
plt.subplot(122);plt.imshow(new_img , cmap='gray');plt.title('Output')
```

```python
_ , thresh1 = cv2.threshold(img , 127 , 255 ,cv2.THRESH_BINARY)
_ , thresh2 = cv2.threshold(img , 127 , 255 ,cv2.THRESH_BINARY_INV)
_ , thresh3 = cv2.threshold(img , 127 , 255 ,cv2.THRESH_TRUNC)
_ , thresh4 = cv2.threshold(img , 127 , 255 ,cv2.THRESH_TOZERO)
_ , thresh5 = cv2.threshold(img , 127 , 255 ,cv2.THRESH_TOZERO_INV)

plt.figure(figsize=(10 , 10))
plt.subplot(231);plt.imshow(thresh1 , cmap='gray');plt.title('THRESH_BINARY')
plt.subplot(232);plt.imshow(thresh2 , cmap='gray');plt.title('THRESH_BINARY_INV')
plt.subplot(233);plt.imshow(thresh3 , cmap='gray');plt.title('THRESH_TRUNC')
plt.subplot(234);plt.imshow(thresh4 , cmap='gray');plt.title('THRESH_TOZERO')
plt.subplot(235);plt.imshow(thresh5 , cmap='gray');plt.title('THRESH_TOZERO_INV')
plt.subplot(236);plt.imshow(img , cmap='gray');plt.title('Original')
```
