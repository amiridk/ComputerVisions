### Otsu's method, developed by Nobuyuki Otsu in 1979, is a widely used technique in image processing for automatic image thresholding. It is particularly effective for segmenting an image into foreground and background by determining an optimal threshold value that minimizes intra-class variance or, equivalently, maximizes inter-class variance between the two groups of pixels. The method operates on the histogram of the image, analyzing the distribution of pixel intensities to find the threshold that best separates the classes.
> Otsu's algorithm is non-parametric and unsupervised, meaning it does not require prior knowledge about the image or manual intervention. Due to its simplicity and efficiency, it is commonly applied in tasks such as object detection, edge detection, and image binarization, making it a fundamental tool in computer vision and image analysis.

```python
# Thresholding with numpy 

import cv2 
import numpy as np 
import matplotlib.pyplot as plt

img = cv2.imread('../data/images/noisy.jpg')
img = cv2.cvtColor(img , cv2.COLOR_BGR2GRAY)
_ , thresh1 = cv2.threshold(img , 127,255 , cv2.THRESH_BINARY)
_ , thresh2 = cv2.threshold(img , 0 , 255 , cv2.THRESH_BINARY+cv2.THRESH_OTSU)

blur = cv2.GaussianBlur(img , (5,5) , 0)
_ , thresh3 = cv2.threshold(blur, 0 ,255 , cv2.THRESH_BINARY + cv2.THRESH_OTSU)

images = [img, 0, thresh1,
          img, 0, thresh2,
          blur, 0, thresh3]
titles = ['Original Noisy Image','Histogram','Global Thresholding (v=127)',
          'Original Noisy Image','Histogram',"Otsu's Thresholding",
          'Gaussian filtered Image','Histogram',"Otsu's Thresholding"]

plt.figure(figsize=[15,7])
for i in range(3):
    #first column
    plt.subplot(3,3,i*3+1),plt.imshow(images[i*3],'gray'),
    plt.title(titles[i*3]), plt.xticks([]), plt.yticks([])
    #second column
    plt.subplot(3,3,i*3+2),plt.hist(images[i*3].ravel(),256)
    plt.title(titles[i*3+1]), plt.xticks([]), plt.yticks([])
    #third column
    plt.subplot(3,3,i*3+3),plt.imshow(images[i*3+2],'gray')
    plt.title(titles[i*3+2]), plt.xticks([]), plt.yticks([])
plt.show()
```
