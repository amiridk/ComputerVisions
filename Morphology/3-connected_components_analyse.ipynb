### Connected components analysis in image processing is a technique used to identify and label distinct regions or objects within a binary image. It works by scanning the image and grouping pixels that are connected based on a predefined connectivity criterion, such as 4-connected or 8-connected neighborhoods. 
> Each group of connected pixels is assigned a unique label, allowing for the segmentation and analysis of individual objects. This method is widely used in applications like object detection, image segmentation, and feature extraction, as it helps isolate regions of interest for further processing. Connected components analysis is particularly useful in tasks such as counting objects, measuring their properties, or tracking them across frames in video sequences.

```python
import numpy as np
import cv2 
import matplotlib.pyplot as plt
```

```python

img = cv2.imread('../data/images/shapes.png')

img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
_ , img = cv2.threshold(img , 127 , 255 , cv2.THRESH_BINARY)

num_labels ,labels = cv2.connectedComponents(img)

label_hue = np.uint8(179*labels/np.max(labels))
blank = np.ones_like(img)*255
labeled_img = cv2.merge([label_hue, blank, blank])

labeled_img = cv2.cvtColor(labeled_img, cv2.COLOR_HSV2BGR)
labeled_img[label_hue == 0 ] = 0

plt.figure(figsize=(12,5))

plt.subplot(121)
plt.imshow(img, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(122) 
plt.imshow(labeled_img)
plt.title('Labeled Components')
plt.axis('off')

plt.tight_layout()
plt.show()
```
