```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread('../data/images/transparency.png' , cv2.IMREAD_UNCHANGED)
img.shape
```

```python

bgr_image = img[...,0:3]
alpha_channel  = img[...,-1]
rgb_image = cv2.cvtColor(bgr_image, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_image);plt.title('RGB')
```

```python
plt.imshow(alpha_channel , cmap='gray');plt.title('Alpha channel')
```

```python
img = cv2.imread('../data/images/input.jpg' , cv2.IMREAD_UNCHANGED)
plt.imshow(img[... , ::-1] , cmap='gray')
img.shape
```

```python
circle = np.zeros((600 ,800,3) , np.uint8)
cv2.circle(circle, (400, 310), 250, (255, 255, 255), -1)
out_img = cv2.bitwise_and(img, circle)
plt.imshow(out_img[... , ::-1], cmap='gray')
```
