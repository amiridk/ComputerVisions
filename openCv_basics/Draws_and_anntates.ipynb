```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image_blacked = np.zeros((512 , 512 ,3) , np.uint8)
image_bw = np.zeros((512,512) , np.uint8)


plt.imshow(image_blacked[... , ::-1] , cmap='gray')
plt.figure()
plt.imshow(image_bw, cmap='gray')
```

```python
image_blacked = np.zeros((512 , 512 ,3) , np.uint8)

cv2.line(image_blacked , (0,0) , (511,511) , (255,127,0) , 5)

plt.imshow(image_blacked[... , ::-1] , cmap='gray')
```

```python
image_blacked = np.zeros((512 , 512 ,3) , np.uint8)

cv2.rectangle(image_blacked , (200, 200), (300, 300), (255, 0, 0), 4 )

plt.imshow(image_blacked[... , ::-1] , cmap='gray')
```
