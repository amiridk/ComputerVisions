### In image processing, contours are used to detect and represent the boundaries of objects within an image. They are essentially curves that join continuous points along the edges of objects, sharing the same color or intensity. Contours are particularly useful in object detection, shape analysis, and image segmentation tasks.
>  By extracting contours, algorithms can identify and isolate objects, measure their properties (such as area, perimeter, or centroid), and even recognize shapes. Techniques like edge detection (e.g., using the Canny edge detector) are often employed to find these boundaries before contour extraction. Once detected, contours can be approximated, filtered, or used to create hierarchical representations, enabling advanced applications like motion tracking, object recognition, and computer vision tasks. OpenCV, a popular image processing library, provides robust tools for contour detection and manipulation, making it a key component in many computer vision pipelines.

```python
import numpy as np 
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('../data/images/shapes.png')
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
_,thresh = cv2.threshold(img, 127 ,255 ,0)
contours , hierarchy  = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
```

```python
for c in contours:
    print(c.shape)
```

```python
bar_image = cv2.cvtColor(img , cv2.COLOR_GRAY2BGR)
cv2.polylines(bar_image , [contours[0]] , True,(0,0,255),5 )
plt.imshow(bar_image[...,::-1])
```

```python
bar_image = cv2.cvtColor(img , cv2.COLOR_GRAY2BGR)
for c in contours:
    cv2.polylines(bar_image , [c] , True,(0,0,255),5 )
    plt.imshow(bar_image[...,::-1])
```

```python
import numpy as np 
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('../data/images/shapes.png')
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
_,thresh = cv2.threshold(img, 127 ,255 ,0)
contours , hierarchy  = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
bgr_img = cv2.cvtColor(img, cv2.COLOR_GRAY2BGR)
all_contours = cv2.drawContours(bgr_img, contours, -1, (0,255,0), 3)
plt.imshow(all_contours)
```
