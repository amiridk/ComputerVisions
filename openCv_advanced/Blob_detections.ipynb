```python
import numpy as np 
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('../data/images/blob.jpg')

detector = cv2.SimpleBlobDetector_create()
keypoints = detector.detect(img)

result1 = cv2.drawKeypoints(img, keypoints, np.array([]), (0,0,255))
result2 = cv2.drawKeypoints(img, keypoints, np.array([]), (0, 0, 255), cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

plt.figure(figsize=(15,15))

plt.subplot(131);plt.imshow(img[...,::-1]);plt.title('Original Image')
plt.subplot(132);plt.imshow(result1[..., ::-1]);plt.title('Without Rich Keypoints')
plt.subplot(133);plt.imshow(result2[..., ::-1]);plt.title('With Rich Keypoints')
```
