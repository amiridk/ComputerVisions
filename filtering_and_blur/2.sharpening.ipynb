```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('../data/images/plate.jpg')
kernel_1 = np.array([[0,0,0],
                      [0,2,0],
                      [0,0,0]])

kernel_2 = 1/9 * np.array([[1,1,1],
                            [1,1,1],
                            [1,1,1]])

final_kernel = kernel_1 - kernel_2

sharped = cv2.filter2D(img, -1, final_kernel)

plt.figure(figsize=(12,6))

plt.subplot(121)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

plt.subplot(122)
plt.imshow(cv2.cvtColor(sharped, cv2.COLOR_BGR2RGB))
plt.title('Sharpened Image')
plt.axis('off')

plt.tight_layout()
plt.show()
```
