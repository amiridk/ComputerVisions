```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('../data/images/gradient.png' , 0)
kernel = np.array([[0,0,0],
                   [-1/2 , 0 , 1/2],
                   [0,0,0]
                   ])
edges = cv2.filter2D(img, -1, kernel)

plt.figure(figsize=(12,4))

plt.subplot(121)
plt.imshow(img, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(122) 
plt.imshow(edges, cmap='gray')
plt.title('Edge Detection')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
img = cv2.imread('../data/images/little_girl.jpg' , 0)
kernel1 = np.array([[0,0,0],
                   [-1/2 , 0 , 1/2],
                   [0,0,0]
                   ])
kernel2 = np.array([[0,0,0],
                   [1/2 , 0 , -1/2], 
                   [0,0,0]
                   ])
edges1 = cv2.filter2D(img, -1, kernel1)
edges2 = cv2.filter2D(img, -1, kernel2)

plt.figure(figsize=(15,4))

plt.subplot(131)
plt.imshow(img, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(132) 
plt.imshow(edges1, cmap='gray')
plt.title('Edge Detection (Kernel 1)')
plt.axis('off')

plt.subplot(133)
plt.imshow(edges2, cmap='gray')
plt.title('Edge Detection (Kernel 2)')
plt.axis('off')

plt.tight_layout()
plt.show()
```
