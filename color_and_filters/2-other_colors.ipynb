### In image processing, colors play a crucial role in analyzing and manipulating visual data. Images are typically represented using color models such as RGB (Red, Green, Blue), where each pixel is defined by its intensity in these three primary colors. Other models like HSV (Hue, Saturation, Value) and CMYK (Cyan, Magenta, Yellow, Key/Black) are also used depending on the application.
>  Color information helps in tasks like object detection, segmentation, and enhancement. For instance, in medical imaging, specific color ranges can highlight abnormalities, while in computer vision, color-based filtering can isolate objects of interest. Advanced techniques, such as color space transformations and histogram equalization, are employed to improve image quality and extract meaningful information. Understanding and manipulating colors is essential for achieving accurate and visually appealing results in image processing.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img_bright = cv2.imread('../data/images/cube-bright.png')
img_dark = cv2.imread('../data/images/cube-dark.png')

plt.figure(figsize=[11,11])
plt.subplot(331);plt.imshow(img_bright[...,::-1]);plt.title("bright");
plt.subplot(332);plt.imshow(img_dark[...,::-1]);plt.title("dark");

plt.subplot(334);plt.imshow(img_bright[...,0], cmap='gray');plt.title("brigh-blue");
plt.subplot(335);plt.imshow(img_bright[...,1], cmap='gray');plt.title("brigh-green");
plt.subplot(336);plt.imshow(img_bright[...,2], cmap='gray');plt.title("brigh-red");

plt.subplot(337);plt.imshow(img_dark[...,0], cmap='gray');plt.title("dark-blue");
plt.subplot(338);plt.imshow(img_dark[...,1], cmap='gray');plt.title("dark-green");
plt.subplot(339);plt.imshow(img_dark[...,2], cmap='gray');plt.title("dark-red");
```

```python
# YCrCb
import cv2
import numpy as np
import matplotlib.pyplot as plt

img_bright = cv2.imread('../data/images/cube-bright.png')
img_dark = cv2.imread('../data/images/cube-dark.png')

brightYCrCb = cv2.cvtColor(img_bright, cv2.COLOR_BGR2YCrCb)
darkYCrCb = cv2.cvtColor(img_dark, cv2.COLOR_BGR2YCrCb)

#show
plt.figure(figsize=[11,11])
plt.subplot(331);plt.imshow(img_bright[...,::-1]);plt.title("bright");
plt.subplot(332);plt.imshow(img_dark[...,::-1]);plt.title("dark");

plt.subplot(334);plt.imshow(brightYCrCb[...,0], cmap='gray');plt.title("brigh-Y");
plt.subplot(335);plt.imshow(brightYCrCb[...,1], cmap='gray');plt.title("brigh-Cr");
plt.subplot(336);plt.imshow(brightYCrCb[...,2], cmap='gray');plt.title("brigh-Cb");

plt.subplot(337);plt.imshow(darkYCrCb[...,0], cmap='gray');plt.title("dark-Y");
plt.subplot(338);plt.imshow(darkYCrCb[...,1], cmap='gray');plt.title("dark-Cr");
plt.subplot(339);plt.imshow(darkYCrCb[...,2], cmap='gray');plt.title("dark-Cb");
```
